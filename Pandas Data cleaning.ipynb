```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame({'From_To': ['LoNDon_paris', 'MAdrid_miLAN','londON_StockhOlm','Budapest_PaRis', 'Brussels_londOn'],
'FlightNumber': [10045, np.nan, 10065, np.nan, 10085],
'RecentDelays': [[23, 47], [], [24, 43, 87], [13], [67, 32]],
'Airline': ['KLM(!)', '<Air France> (12)', '(British Airways. )','12. Air France', '"Swiss Air"']})
```

```python
df
```

```python
# Question 1

for i, null in enumerate(df['FlightNumber'].isnull()):
    if null:
        df.loc[i,'FlightNumber'] = df.loc[i-1,'FlightNumber']+10
        
df['FlightNumber'] = df['FlightNumber'].astype('int64') 
df
```

```python
# Question 2

from_ = []
to = []
for i in df['From_To'].apply(lambda x: x.split('_')):
    from_.append(i[0])
    to.append(i[1])
temp = pd.DataFrame()
temp['From'] = from_
temp['To'] = to


temp
```

```python
# Question 3

temp['From'] = temp['From'].str.capitalize()
temp['To'] = temp['To'].str.capitalize()

temp
```

```python
# Question 4

df.drop('From_To', axis=1, inplace=True)
df = df.join(temp )

df
```

```python
# Question 5

delays = pd.DataFrame(data=np.nan, index = df.index, columns=['delay_1','delay_2','delay_3'])
for row, lst in enumerate(df['RecentDelays']):
    for ind, val in enumerate(lst):
        delays.loc[row, 'delay'+'_'+str(ind+1)] = val
        
df= df.drop('RecentDelays', axis=1)
df =df.join(delays)

df
```
